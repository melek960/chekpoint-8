```python
import pandas as pd 
import numpy as np 
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    
df=pd.DataFrame(te_ary, columns=te.columns_)  
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) 
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
data= pd.read_table("C:\\Users\\melek\\Downloads\\Market_Basket_Optimisation.csv", delimiter =",", header= None)  
```

```python
data.info()
```

```python
listt=[]
for j,row in data.iterrows():
    l=[]
    for i in row:
        if  type(i)==str:
            l.append(i)
    listt.append(l)        
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(listt).transform(listt)    
df=pd.DataFrame(te_ary, columns=te.columns_)  
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.005)
```

```python
frequent_itemsets=apriori(df, min_support=0.05, use_colnames=True) 
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.07)
```

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```
